```python
import numpy as np
import random as Random
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
```

```python
# Kmeans Function wrapper
def KMEANS(X,k=3):
    kmeans = KMeans(n_clusters= k)
 
    #predict the labels of clusters.
    labels = kmeans.fit_predict(X)
    return(labels)
```

```python
# Generate sample data
N = 750     # No of points
centers = [[1, 1], [-1, -1], [1, -1]]
X, cluster_labels = make_blobs(n_samples=N, centers=centers, cluster_std=0.4,
                            random_state=0)
print(X)
X = StandardScaler().fit_transform(X)
print(X)
```

```python
xcord = [X[i][0] for i in range(N)]
ycord = [X[i][1] for i in range(N)]
plt.scatter(xcord,ycord)
```

```python
label = KMEANS(X)

filtered_label0 = X[label == 0]
filtered_label1 = X[label == 1]
filtered_label2 = X[label == 2]

#plotting the results
plt.scatter(filtered_label0[:,0] , filtered_label0[:,1], color = 'red')
plt.scatter(filtered_label1[:,0] , filtered_label1[:,1], color = 'blue')
plt.scatter(filtered_label2[:,0] , filtered_label2[:,1], color = 'green')
plt.show()
```
