```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# (a)
m = np.zeros(2)
c = np.eye(2)

x,y = np.random.multivariate_normal(m,c,100).T

plt.scatter(x,y)
plt.axis('equal')
plt.show()
plt.clf()
```

```python
# (b)
x,y = np.random.multivariate_normal([-1,1],c,100).T

plt.scatter(x,y)
plt.axis('equal')
plt.show()
plt.clf()
```

```python
# (c)

x,y = np.random.multivariate_normal(m,2*c,100).T

plt.scatter(x,y)
plt.axis('equal')
plt.show()
plt.clf()
```

```python
# (d)
c = [[1,0.5],[0.5,1]]
x,y = np.random.multivariate_normal(m,c,100).T

plt.scatter(x,y)
plt.axis('equal')
plt.show()
plt.clf()
```

```python
# (e)
c = [[1,-0.5],[-0.5,1]]
x,y = np.random.multivariate_normal(m,c,100).T

plt.scatter(x,y)
plt.axis('equal')
plt.show()
plt.clf()
```
