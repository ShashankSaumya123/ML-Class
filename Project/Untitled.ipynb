```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imshow
from skimage.transform import rescale
from os import listdir
```

```python
# load all images in a directory
X = []
Y = []
for filename in listdir('C:/Users/DELL/Desktop/NISER/ML-Class/Project/WBC_data_kaggle/main_dataset/test/EOSINOPHIL'):
    # load image
    img_data = imread('C:/Users/DELL/Desktop/NISER/ML-Class/Project/WBC_data_kaggle/main_dataset/test/EOSINOPHIL/' + filename)
    im_data = img_data[:,:,2]
    #Rescaling the image
    image_r = rescale(im_data, 0.4, anti_aliasing=False)
    X.append(image_r)
    Y.append(0)
```

```python
plt.imshow(X[5])
plt.title('Aliasing')
#plt.savefig('Aliased')
plt.show()
plt.clf()
```
